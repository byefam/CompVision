# 1.4.3 Geometric Transformations of Images

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

## Scaling

```python
img = cv2.imread('dsu2.jpg')
res = cv2.resize(img,None,fx=2, fy=2, interpolation = cv2.INTER_NEAREST)
res = cv2.resize(img,None,fx=2, fy=2, interpolation = cv2.INTER_CUBIC)
```

```python
cv2.imshow('image',img)
cv2.imshow('res image',res)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
height, width = img.shape[:2]
res = cv2.resize(img,(2*width, 2*height), interpolation = cv2.INTER_CUBIC)
cv2.imshow('image',img)
cv2.imshow('res image',res)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Translation

```python
img = cv2.imread('dsu4.jpg',1)
tx = 300
ty = 150
rows,cols,ch = img.shape
M = np.float32([[1,0,tx],[0,1,ty]])
dst = cv2.warpAffine(img,M,(cols,rows))
dst = cv2.rectangle(dst,(0,0),(tx,ty),(0,255,0),-1)
cv2.imshow('img',dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Rotation

```python
img = cv2.imread('dsu3.jpg',1)
rows,cols,ch = img.shape
cx = 100
cy = 0
theta = 45
M = cv2.getRotationMatrix2D((cy,cx),theta,1)
dst = cv2.warpAffine(img,M,(cols,rows))
img = cv2.circle(dst,(cx,cy), 63, (0,0,255), -1)
cv2.imshow('img',dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Affine Transformation

```python
img = cv2.imread('dsu2.jpg')
rows,cols,ch = img.shape
pts1 = np.float32([[235,457],[238,269],[569,231]])
pts2 = np.float32([[235,457],[235,270],[570,270]])

M = cv2.getAffineTransform(pts1,pts2)

dst = cv2.warpAffine(img,M,(cols,rows))

plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```

```python
cv2.imshow('img',dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Perspective Transformation

```python
img = cv2.imread('dsu1.jpg')
rows,cols,ch = img.shape

pts1 = np.float32([[579,338],[753,375],[763,543],[584,543]])
pts2 = np.float32([[579,340],[760,340],[760,540],[579,540]])
M = cv2.getPerspectiveTransform(pts1,pts2)
dst = cv2.warpPerspective(img,M,(cols,rows))

plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```
